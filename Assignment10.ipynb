#Glass Classification


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read train and test sets
train = pd.read_csv("trainKNN.txt", header=None)
train.columns = ['ID', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type of glass']
test = pd.read_csv("testKNN.txt", header=None)
test.columns = ['ID', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type of glass']
```

```python
train.head(10)
```

```python
test.head()
```

from the above data we can drop the ID column from both the train and test datasets


```python
train = train.drop('ID',axis = 1)
test = test.drop('ID',axis = 1)
```

```python
train.describe()
```

```python
test.describe()
```

```python
train.info()
```

```python
test.info()
```

standardizing the columns in both datasets expect the target column

```python
def standardize (df):
    for col in df.columns:
        if col != "Type of glass": # Don't standardize the categories
            df[col] = (df[col] - df[col].mean())/df[col].std()
    return df
```

```python
# Standardize data
train = standardize(train)
test = standardize(test)
```

```python
train.head()
```

splitting the data in X_train,X_test,y_train,y_test

```python
#splitting train data
y_train = train['Type of glass']
X_train = train.drop('Type of glass',axis =1)

#splitting test data
y_test = test['Type of glass']
X_test = test.drop('Type of glass',axis =1)
```

Building the models:

1. Square Euclidean Distance Model

2. Manhattan Distance Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from scipy.spatial import distance
knc_euclidean_model = KNeighborsClassifier(n_neighbors = 8, metric = distance.sqeuclidean)# Square Euclidean distance model
knc_manhattan_model = KNeighborsClassifier(n_neighbors = 8, metric = distance.cityblock )# Manhattan distance model
```

Training the models

```python
knc_euclidean_model.fit(X_train,y_train)
knc_manhattan_model.fit(X_train,y_train)
```

```python
# Make predictions
euclid_predictions = knc_euclidean_model.predict(X_test) 
manhattan_predictions = knc_manhattan_model.predict(X_test)
df = pd.DataFrame({'actual': y_test, 'euclid': euclid_predictions, 'manhattan': manhattan_predictions})
df.head()
```

```python
from sklearn.metrics import classification_report
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
# Evaluate performance
manhattan_count = len(df.loc[df['manhattan'] == df['actual']])
euclid_count = len(df.loc[df['euclid'] == df['actual']])
print('Manhattan Accuracy: {}%'.format(round(100*manhattan_count/len(df), 2)))
print(classification_report(y_test, manhattan_predictions, target_names=df['actual'].astype(str).unique()))
print ('\n')
print('Square Euclidean Accuracy: {}%'.format(round(100*euclid_count/len(df), 2)))
print(classification_report(y_test, euclid_predictions, target_names=df['actual'].astype(str).unique()))
```

**The performance of Manhattan Model Accuracy: 66.67% is good compare with the performance of Square Euclidean Model Accuracy: 61.11%.**
